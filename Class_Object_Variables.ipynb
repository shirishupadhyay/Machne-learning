```python
class Car:
  def __init__(self,windows,doors,enginetype):
    self.windows = windows
    self.doors = doors
    self.enginetype = enginetype
  def self_driving(self):
    return "This is a {} car".format(self.enginetype)
```

```python
car1 = Car(4,5,"petrol")
```

```python
car1.self_driving()
```

```python
print(car1.windows)
print(car1.enginetype)
print(car1.doors)
```

```python
car2 = Car(10, 20, "Diesel")
```

```python
car2.self_driving()
```

```python
print(car2.windows)
print(car2.doors)
print(car2.enginetype)
```
