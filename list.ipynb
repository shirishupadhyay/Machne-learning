# **List Comprehension**

```python
lst1 =[]
def list_square(lst):
  for i in lst:
    lst1.append(i*i)
  return lst1
```

```python
list_square([1,2,3,4,5,6,7,8,9,0])
```

```python
lst = [2,4,6,7,8,9]
```

```python
lst1 = [i*i for i in lst if i%2 ==0]
```

```python
print(lst1)
```
