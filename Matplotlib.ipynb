```python
#sine curve
x = np.arange(0, 4* np.pi, 0.1)
y = np.sin(x)
plt.title("Sine Wave Form")

plt.plot(x,y)
plt.show()
```

```python
#subplot

x = np.arange(0, 5*np.pi, 0.1)
y_sin= np.sin(x)
y_cos = np.cos(x)


#first plot
plt.subplot(2,2,1)
plt.plot(x,y_sin, "r*-")
plt.title("Sine")

#second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,"g*-")
plt.title("Cosine")

plt.show()
```

# **Bar Plot**

```python
#bar plot
x1 = [10,20,30]
y1 = [21,31,41]

x2 = [15,25,35]
y2=[26,36,46]

plt.bar(x1,y1, color ="r")
plt.bar(x2,y2,color ="b")

plt.title("Bar Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()
```

# **Histogram**

```python
a=np.array([22,43,98,76,54,32,45,12,11,5,90,100])
plt.hist(a)
plt.title("Histogram")
plt.show()
```

# **Box Plot**

```python
data = [np.random.normal(0, std, 100) for std in range (1,4)]
plt.boxplot(data, vert=False, patch_artist = False)
plt.show()
```

# **Pie Chart**

```python
#data to plot
labels = "Python", "C++", "Rust", "JavaScript"
sizes = [215, 150, 175, 200]
colors = ["gold", "lightskyblue", "yellowgreen", "lightcoral"]

explode = (0.12, 0,0, 0)
plt.pie(sizes,explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True)
plt.axis("equal")
plt.show()
```
