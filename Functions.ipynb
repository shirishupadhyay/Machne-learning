**List**

```python
lst = [1 ,4 ,5 ,6 ,7 ,8]
```

```python
def even_odd(lst):
  even_sum = 0
  odd_sum = 0
  for i in lst:
    if i%2 == 0:
      even_sum = even_sum +i
    else:
      odd_sum = odd_sum + i
  return even_sum, odd_sum
```

```python
even_odd(lst)
```

# **Map Function**

```python
def even_or_odd(num):
  if num%2 == 0:
    print(f"The number {num} is even")
  else:
    print(f"The number {num} is odd")
```

```python
even_or_odd(10)
```

```python
lst = [2, 3, 4, 5, 6, 7, 8, 9]
```

```python
list(map(even_or_odd ,lst ))
```

# **Filter Function**

```python
def even(num):
  if num%2 ==0:
    return True
```

```python
lst =[1,2,3,4,5,6,7,8,9]
```

```python
list(filter(even, lst))
```

```python
list(filter(lambda num:num%2==0, lst))
```

```python
list(map(lambda num:num%2 == 0, lst))
```
